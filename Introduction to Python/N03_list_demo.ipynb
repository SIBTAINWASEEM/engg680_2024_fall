```python
# Create a list of integers
numbers = [1, 2, 3, 4, 5]
print(numbers)

# Create a list of different data types
mixed_list = [1, "Hello", 3.14, True]
print(mixed_list)
```

```python
# Access elements by index
print("The first element:", mixed_list[0])  # Output: 1
# Modify elements by index
mixed_list[0] = "World"  # Change 1 to "World"
print("The second element after change is :", mixed_list[0])
```

```python
# Slice from index 1 to 4 (not including 4)
slice_of_numbers = numbers[1:4]

# Slice from the beginning up to index 3
begin_to_three = numbers[:3]

# Slice from index 2 to the end
from_two_to_end = numbers[2:]

print(slice_of_numbers)
print(begin_to_three)
print(from_two_to_end)
```

```python
# Append an element to the end of the list
numbers.append(6)

# Extend the list by appending all the items from another list
numbers.extend([7, 8])

# Insert an element at a specific position
numbers.insert(0, 0)  # Inserts at the beginning

print(numbers)
```

```python
# Remove the first occurrence of an element
numbers.remove(5)

# Pop the last item and return it
last_item = numbers.pop()

# Remove an element at a specific position
del numbers[0]  # Deletes the first element

print(numbers)
```

```python
# Create a list of squares from another list
squares = [x**2 for x in numbers]

# Filter the list to include only even numbers
evens = [x for x in numbers if x % 2 == 0]

print(squares)
print(evens)
```

```python
# Create two lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Concatenate the lists
combined_list = list1 + list2

print(combined_list)  # Output will be [1, 2, 3, 4, 5, 6]
```

```python
fruits = ['apple', 'banana', 'cherry', 'date']

# Using a simple for loop to traverse the list
for fruit in fruits:
    print(fruit)

# Using enumerate to get both index and value
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# Traverse with a while loop
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
```
